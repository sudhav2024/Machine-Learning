```python
df1 = pd.DataFrame({'A': [1,2,3],'B': [4,5,6]})
df2 = pd.DataFrame({'A': [3,4,5],'B': [6,7,8]})
print(df1)
print(df2)
print("Addition:\n",df1+df2)
print("\nSubtraction:\n",df1-df2)
print("\nMultiplication:\n",df1*df2)
print("\nDivision:\n",df1/df2)
```

```python
df3 = pd.DataFrame({'A': [1,2,3],'B': [4,5,6]}, index=[0,1,2])
df4 = pd.DataFrame({'A': [3,4,5],'B': [6,7,8]}, index=[1,2,3])
print(df3)
print(df4)
result = df3 + df4
print("Result of df3 + df4 with aligment:\n", result)
```
